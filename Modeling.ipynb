```python
import pandas as pd
import os
import ast
import numpy as np
import pickle
from datetime import datetime
```

```python
path = 'C:/Users/HEN1/Projects/YELP_DATASET/DATA'
file_name = 'yelp.pkl'
file = os.path.join(path, file_name)
with open(file, 'rb') as file:
    df=pickle.load(file)
```

```python
df = df.drop(columns=['date', 'yelping_since'])
```

```python
df[df['review_count']<10].shape[0]
```

```python
df.head()
```

```python
trans_label = {1:0, 2:1, 3:2, 4:3, 5:4}
df['stars'] = df['stars'].map(trans_label)
```

```python
df = df.loc[:,df.apply(pd.Series.nunique) != 1]
```

```python
for col in df.columns.values.tolist():     
    print({col:df[col].unique()})
```

```python
# check lightgbm version
import lightgbm as lgb
print(lgb.__version__)
```

```python
df.shape
```

```python
x = df.drop(columns='stars')
y = df.stars
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
x_train.head()
```

## Light GBM

```python
train_data=lgb.Dataset(x_train,label=y_train)
```

```python
param = {'objective': 'multiclass', 'learning_rate':.01, 'num_class':5, 'num_leaves':40, 'max_depth':6, 'force_col_wise':True}
param['metric'] = ['multi_logloss', 'multi_error']
```

```python
num_round=100
```

```python
start=datetime.now()
lgbm=lgb.train(param,train_data,num_round)
stop=datetime.now()
```

```python
ypred_lgbm=lgbm.predict(x_test)
```

```python
ypred_lgbm = [np.argmax(line) for line in ypred_lgbm]
```

```python
from sklearn.metrics import confusion_matrix, balanced_accuracy_score
print(confusion_matrix(y_test, ypred_lgbm))
print(balanced_accuracy_score(y_test, ypred_lgbm))
```

```python
ypred_lgbm_train = lgbm.predict(x_train)
ypred_lgbm_train = [np.argmax(line) for line in ypred_lgbm_train]
print(balanced_accuracy_score(y_train, ypred_lgbm_train))
```

## Naive Bayes

```python
df_nostr = df.drop(columns=['user_id', 'business_id', 'postal_code'])
```

```python
xn = df_nostr.drop(columns='stars')
yn = df_nostr.stars
```

```python
from sklearn.model_selection import train_test_split 
xn_train,xn_test,yn_train,yn_test=train_test_split(xn,yn,test_size=.3)
```

```python
xn_train.head()
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(xn_train, yn_train)
ypred_gnb = gnb.predict(xn_test)
```

```python
# accuracy on X_test
accuracy = gnb.score(xn_test, yn_test)
print(f"accuracy for baive bayes: {accuracy}")
 
# creating a confusion matrix
cm_nb = confusion_matrix(yn_test, ypred_gnb)
print(cm_nb)
```
